```python
df1 = df.drop_duplicates()
df1 = df.reset_index(drop=True)
```

```python
from konlpy.tag import Twitter

twitter = Twitter()
```

```python
test = df['txt'][1]
```

```python
test
```

```python
morphs_list = []

for text in df1['txt']:
    # Apply twitter.morphs() and append the result to the list
    morphs_list.append(twitter.morphs(text))
```

```python
len(morphs_list)
```

```python
from collections import Counter
from itertools import chain

# Flatten the morphs_list
all_morphs = list(chain.from_iterable(morphs_list))

# Create the Counter
count = Counter(all_morphs)

# Order the Counter (most frequent to least frequent)
ordered_count = count.most_common()

ordered_count
```

```python
import csv

# ... (Previous code to create ordered_count) ...

# Specify the CSV file path
csv_file_path = 'morpheme_counts.csv'

# Write the data to the CSV file
with open(csv_file_path, 'w', newline='', encoding='utf-8') as csvfile:
    writer = csv.writer(csvfile)

    # Write the header row
    writer.writerow(['Morpheme', 'Count'])

    # Write the data rows
    writer.writerows(ordered_count)
```
